### ðŸŽ¯ Your Task:
Using the existing DataFrame, perform the following operations:

âœ… Print the details (name, age, marks) of the student who scored the highest marks.

âœ… Add a new student "Tharushi" with age 23 and marks 88 using the index "Th".

âœ… Display the details of students who are 25 years old or older.

âœ… Increase the marks of students below 80 by 10 bonus points.

âœ… Calculate and print the average marks of all students after the bonus.


```python
import pandas as pd
ser1 = pd.Series({'ru':'rukmini','vs':'vasanthan','pr':'priyanka','mo':'mohan'})
ser2 = pd.Series({'ru':23,'vs':24,'pr':54,'mo':43})
ser3 = pd.Series({'ru':97,'vs':99,'pr': 56,'mo':65})
cols = {
    'name':ser1,
    'age':ser2,
    'marks':ser3
}
df = pd.DataFrame(cols)

# Print the details (name, age, marks) of the student who scored the highest marks.
highest_marks_student = df.loc[df['marks'].idxmax()]
print("Student with the highest marks:")
print(highest_marks_student)

# Add a new student "Tharushi" with age 23 and marks 88 using the index "Th".
df.loc['Th'] = pd.Series({'name':'Tharushi','age':23,'marks':88})
print("\nDataFrame after adding Tharushi:")
print(df)

# Display the details of students who are 25 years old or older.
print("\nStudents 25 years old or older:")
print(df[df['age'] >= 25])

# Increase the marks of students below 80 by 10 bonus points.
df.loc[df['marks'] < 80, 'marks'] += 10
print("\nDataFrame after adding bonus marks:")
print(df)

# Calculate and print the average marks of all students after the bonus.
average_marks = df['marks'].mean()
print(f"\nAverage marks of all students after bonus: {average_marks}")
```

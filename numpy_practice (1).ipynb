```python
import numpy as np
a=np.array([[10,12,11],
            [9,11,10],
            [12,13,14],
            [8,7,9]])
b=np.array([[20,19,21],
            [18,20,19],
            [22,23,21],
            [17,18,16]])
c=np.array([[15,16,14],
            [14,13,15],
            [16,17,18],
            [12,11,13]])
production_all=np.array([a,b,c], dtype=float)
print(production_all.shape)
print(production_all.size)
print(production_all.dtype)
print(production_all.ndim)
print("sum for farm1",np.sum(production_all[0]))
print("sum for farm2",np.sum(production_all[1]))
print("sum for farm3",np.sum(production_all[2]))
print("mean for farm1",np.mean(production_all[0]))
print("mean for farm2",np.mean(production_all[1]))
print("mean for farm3",np.mean(production_all[2]))
print("std for farm1",np.std(production_all[0]))
print("std for farm2",np.std(production_all[1]))
print("std for farm3",np.std(production_all[2]))
print("var for farm1",np.var(production_all[0]))
print("var for farm2",np.var(production_all[1]))
print("var for farm3",np.var(production_all[2]))
#
hours_weights=np.array([0.5,1.0,0.75])
weighted_per_panel=np.dot(production_all,hours_weights)
print(weighted_per_panel)


production_all[1 :2]+=0.5
print(production_all)
per_farm_mean = np.mean(production_all,axis=(1,2))
per_farm_std = np.std(production_all, axis=(1, 2))
normalized_farm = production_all-per_farm_mean
print(normalized_farm)
normalized_farm=normalized_farm/per_farm_std
print(normalized_farm)
print(np.exp(np.round(normalized_farm)+1))
print(np.expm1(normalized_farm))
print(np.log(np.round(normalized_farm)+1))
print(np.log10(np.round(normalized_farm)+1))
print(np.log2(np.round(normalized_farm)+1))
print(np.degrees(np.sin(normalized_farm)))
print(np.radians(np.sin(normalized_farm)))
print(np.degrees(np.cos(normalized_farm)))
print(np.radians(np.cos(normalized_farm)))
print(np.degrees(np.tan(normalized_farm)))
print(np.radians(np.tan(normalized_farm)))
print(np.degrees(np.arcsin(normalized_farm)))
print(np.degrees(np.arccos(normalized_farm)))
print(np.degrees(np.arctan(normalized_farm)))
print(np.round(normalized_farm))
print(np.floor(normalized_farm))
print(np.ceil(normalized_farm))
print(np.trunc(normalized_farm))

print(np.logical_and(np.greater_equal(normalized_farm,20),production_all%2==0))
print(np.logical_or(production_all>20,production_all%2==0))
print(np.logical_xor(production_all>20,production_all%2==0))
```
